# Estamos adicionando as duas colunas que estavam dividas no na nossa base de dados "endereços_sp" e fazemos uma modificação para deixar todas as letras minúsculas.

```python
endereços_sp["rua"] = endereços_sp["tipo_logr"] + ' ' + endereços_sp["logr_nome"]
endereços_sp["rua"] = endereços_sp["rua"].str.lower().str.strip()
endereços_sp
```

# Aplicamos um Regex na nossa coluna rua para eliminar qualquer tipo de pontuação ou numeros, para deixar as duas base de dados iguais.

```python
dados_vendas["Rua"].str.extract(r'(^[\w ]+)')[:10]
```

# Após aplicarmos o regex, usamos .lower() e .strip() para deixar todas as ruas no mesmo padrão do outro banco de dados.

```python
dados_vendas["apenas_rua"] = dados_vendas["Rua"].str.extract(r'(^[\w ]+)')
dados_vendas["apenas_rua"] = dados_vendas["apenas_rua"].str.lower().str.strip()
dados_vendas.head()
```

# Utilizando geopandas conseguimos ler os dados do censo que baixamos e iremos utilizar a coluna "geometry" que contem latitudes e longitudes.

```python
setor_censo = gpd.read_file('/content/drive/MyDrive/Alura imersao dados/35SEE250GC_SIR.shp')
setor_censo
```

```python
ax = setor_censo['geometry'].plot()
ax.set_title('Setores Censitários')
```

# Utilizando o metodo "Point" com .contains(), conseguimos checar se o banco de dados "setor_censo_sp" contém a nosas latitude e longitude do banco de dados "dados_geo", dessa forma como vamos ver abaixo conseguimos o código de setor do nosso outro banco de dados "setor_censo_sp"

```python
from shapely.geometry import Point
latitude = -23.64
longitude = -46.69

setor_censo_sp = setor_censo[setor_censo.NM_MUNICIP == 'SÃO PAULO']
setor_censo_sp[setor_censo_sp.contains(Point(longitude, latitude))]
```

# Estamos criando uma nova coluna chamada "Point" onde vamos ter latitude e longitude, e utilizando essa coluna criaremos uma nova coluna chamada "setor_censo", onde usaremos o comando .contains() utilizando "Point" como parâmetro para cruzarmos os dados e pegarmos o COD_setor.

```python
dados_geo['Point'] = ""
for i in dados_geo.index:
    dados_geo["Point"][i] = Point(dados_geo['longitude'][i], dados_geo["latitude"][i])

dados_geo['setor_censo'] = dados_geo["Point"].map(
    lambda x: setor_censo_sp.loc[setor_censo_sp.contains(x), 'CD_GEOCODI'].values
).str[0]
dados_geo
```

```python
dados_geo['setor_censo'] = dados_geo['setor_censo'].astype(float)
dados_geo.info()
```
