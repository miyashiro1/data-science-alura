```python
import pandas as pd

url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
```

```python
dados.head()
```

# Quando usamos expand=True, serapamos os elementos por colunas. 



```python
dados["Valor"].str.split(expand = True)
```

# Criamos 3 novas colunas chamadas 'Moeda', 'Valor_anuncio', e 'Tipo_anuncio' para atribuir as 3 novas colunas que foram criadas a partir do split usado acima.

```python
dados[['Moeda', 'Valor_anuncio', 'Tipo_anuncio']] = dados["Valor"].str.split(expand = True)
dados
```

# Na coluna "Tipo_anuncio" tinham casos de aluguel, isso estava bagunçando os valores, por isso removemos todos os /mes e deixando apenas os valores contendo "None" e atribuímos essa mudança para um novo dataframe chamado dados_vendas.

```python
dados_vendas = dados[dados['Tipo_anuncio'].isnull()]
dados_vendas
```

# A coluna 'Valor_anuncio' era uma string, por isso criamos uma nova coluna chamada 'Valor_anuncio_float' e transformamos ela em float e removemos o '.' para não causar nenhum problema.

```python
dados_vendas["Valor_anuncio_float"] = dados_vendas['Valor_anuncio'].str.replace('.','').astype(float)
```

# Display.precision para colocar apenas 2 casas decimais quando usarmos .describe()

```python
pd.set_option("display.precision", 2)
pd.set_option('display.float_format', lambda x: '%.2f' %x)
dados_vendas.describe()
```
